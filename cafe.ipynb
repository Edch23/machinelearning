```python
import pandas as pd
# Cargar desde un archivo CSV
df = pd.read_csv('/content/index.csv')  # Asegúrate de reemplazar 'ruta/al/dataset.csv' con la ruta correcta

# Ver las primeras filas del DataFrame
print(df.head())
```

```python
# Estadísticas descriptivas
stats = df['money'].describe()
print(stats)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizar el histograma de las ventas
plt.figure(figsize=(12, 6))
sns.histplot(df['money'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histograma de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()

# Visualizar el boxplot de las ventas
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, y='money', color='lightgreen')
plt.title('Boxplot de Ventas')
plt.ylabel('Ventas')
plt.grid()
plt.show()
```

```python
# Asegurarse de que la columna de fecha esté en formato datetime
df['date'] = pd.to_datetime(df['date'])

# Agrupar por fecha y sumar las ventas
daily_sales = df.groupby('date')['money'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(daily_sales['date'], daily_sales['money'], marker='o', color='orange')
plt.title('Tendencia de Ventas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='money', y='coffee_name', hue='coffee_name', palette='Set1')
plt.title('Relación entre Ventas y Tipo de Café')
plt.xlabel('Ventas')
plt.ylabel('Tipo de Café')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Histograma
sns.histplot(df['money'], bins=30, color='skyblue', edgecolor='black', alpha=0.6)

# Boxplot
sns.boxplot(data=df, y='money', color='lightgreen', width=0.2)

plt.title('Histograma y Boxplot de Ventas')
plt.xlabel('Ventas')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()
```
